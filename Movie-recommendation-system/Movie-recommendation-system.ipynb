- I want to seperate 'action', 'adventures', 'fantasy' words from genres
- I want to bring genres in this format ['Action', 'Adventure', 'Fantasy', 'Science Fiction']

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

###### Error:- TypeError :  string indices must be integers
- Since it is string. So I want to convert string of list into a list so for that I will use one **ast module** in python and inside ast one function is their **ast.literal_eval()** which will exactly do my work
- This is my task, I want to do data pre-processing. So I can simply create a helper function.
- I can pass loop for iterating values

- Now for cast I want to pick top 3 actors / actress.
- From above I want first 3 dictionaries in that all information of actors / actress are their
- The reason of picking top 3 actors / actress from one movie is all the leading role character names are always shown on top

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter = counter + 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

- Now I want all the names of director inside crew.

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

- One challenge is that I want only those names whose job is director as above

It will became a big blunder. So by removing white spaces SamWorthington or SamMendes will became a one entity and have only one tag i.e. **SamWorthington** & **SamMendes**

- In library of nltk their is one function called stemming
- I will transform **['action', 'actions']** into one stem like **['action', 'action']** similarly **['accept', 'accepted', 'accepts']** into one stem **['accept', 'accept', 'accept']**.

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

- Again I have made hepler function

```python
from sklearn.metrics.pairwise import cosine_similarity
```

- cosine_similarity is nothing but the distance between two vectors. If the distance is minmal it will show 1 else if the distance between two vectors is more it will show 0.08 or 0

- Here is the link to refer cosine_similarity :

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html

```python
similarity = cosine_similarity(vectors)
```

- I have passed whole vectors to cosine_similarity so that each vector will get compared with another vector and will calculate the distance between them.

- I have total 4806 vectors so each vector will calculate its distance with another vector.

- For example 1st movie will get calculated with 4806, 2nd movie will calculate its distance with 4806, similary all movies...

```python
similarity
```

```python
similarity[0]
```

- This is 1st movie which gets calculated his distance with all 4806 movies.
- 1st movie will calculate its distance with 1st movie that is why it is coming out to be 1
- similarly 2nd movie with each movie distance and it is coming out to be 1 with own 2nd movie.
- So basically this metrics diagonal will always be 1. Last movie distance calculate with last will be 1

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    if len(movies_list) == 0:
        print('No recommendation found')
    else:
    
        for i in movies_list:
            print(new_df.iloc[i[0]].title)
```

```python
recommend('The Hateful Eight')
```
